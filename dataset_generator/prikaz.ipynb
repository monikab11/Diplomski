```python
import networkx as nx
import matplotlib.pyplot as plt

# Sadr≈æaj iz graphs_04.txt
# graph6_lines = ["BW", "CF", "D?{", "E?Bw", "F??Fw", "G???F{", "H????B~", "I??????~w"]
# graph6_lines = ["BW", "Bw"]
# graph6_lines = ["CF", "CR", "CN", "Cr", "C^", "C~"]
# graph6_lines = ["D`["]
graph6_lines = ["D?{"]
# graph6_lines = ["D@s", "C^", "CF", "D`[", "DBw", "BW", "DB{", "Bw"]

# Parsiranje i vizualizacija svakog grafa
plt.figure(figsize=(10, 10))
n = 4
for i, graph6_str in enumerate(graph6_lines, start=1):
    # Parsiranje grafa
    G = nx.from_graph6_bytes(bytes(graph6_str.strip(), "ascii"))

    # Podjela u podgrafove za prikaz
    plt.subplot(n, 3, i)
    nx.draw(G, with_labels=True, node_color="lightblue", node_size=700, font_size=10)
    plt.title(f"Graph: {graph6_str}", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
path = "./graph6/graphs_03.txt"
import math

graph6_lines = []
with open(path, 'r') as file:
    for line in file:
        graph6_lines.append(line.replace("\n", ""))

plt.figure(figsize=(8, 5))
n = math.ceil(len(graph6_lines))
for i, graph6_str in enumerate(graph6_lines, start=1):
    G = nx.from_graph6_bytes(bytes(graph6_str.strip(), "ascii"))

    plt.subplot(n, 3, i)
    nx.draw(G, with_labels=True, node_color="lightblue", node_size=700, font_size=10)
    plt.title(f"Graph: {graph6_str}", fontsize=10)

plt.tight_layout()
plt.show()
```
